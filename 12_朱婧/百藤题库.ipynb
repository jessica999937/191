```python
#1)构建两种网页
urls = [("https://pintia.cn/problem-sets/1111652100718116864/problems/type/6","函数题"),
        ("https://pintia.cn/problem-sets/1111652100718116864/problems/type/7","编程题")]
from selenium import webdriver
import re,time
driver=webdriver.Chrome()

def get_html(url):
    driver.get(url)
    time.sleep(5)
    html = driver.page_source
    return html

def extract_info(html,category):
    import re
    results=[]#[[a,b,c,d,e,f,g],[]]
    reobj = re.compile(r'<tr><td class="answerIcon_1du7d"></td><td>([\d\D]*?)</td><td><a href="([\d\D]*?)" class="">([\d\D]*?)</a></td><td>([\d\D]*?)</td><td>([\d\D]*?)</td><td>([\d\D]*?)</td><td>([\d\D]*?)</td></tr>')
    for match in reobj.finditer(html):
        a=match.group(1)#标号
        b=match.group(2)#网址/problem-sets/1111652100718116864/problems/1111653039147499520
        b= "https://pintia.cn" + b
        c=match.group(3)#标题
        d=match.group(4)#分数
        e=match.group(5)#通过数
        f=match.group(6)#提交数
        g=match.group(7)#通过率
        print("标号:",a,"url:",b,"标题:",c,"分数:",d,"通过数:",e,"提交数:",f,"通过率:",g,category)
        results.append([a,b,c,d,e,f,g,category])
    return results

results = []#所有列表页的信息
for t in urls:
    url , category = t
    html = get_html(url)
    li  = extract_info(html,category)
    for elem in li:
        results.append(elem)
    
print("="*10)
print(results)

```

```python
from selenium import webdriver
import re,time
driver=webdriver.Chrome()
def get_html(url):
    driver.get(url)
    time.sleep(5)
    html = driver.page_source
    return html

urls = ["https://pintia.cn/problem-sets/1111652100718116864/problems/1111653039147499521","https://pintia.cn/problem-sets/1111652100718116864/problems/1111653039147499522"]


def get_content(html):
    # print(html)
    result = []
    reobj = re.compile(r'返回[\d\D]*?(<div class="ques-view">[\d\D]*</pre>[\d\D]*?</div>)[\d\D]*?陈春')
    for match in reobj.finditer(html):
        result.append(match.group(1))
    return result    

contents = []
for url in urls:
    print(url)
    html = get_html(url)
    print("-"*10)
    content = get_content(html)
    contents.append(content)
print("="*10)
print(contents)

    
```
