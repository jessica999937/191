# 思路：
1.建立两个网页的urls=[(),()]，循环打开两个url
2.正则信息抽取标号、网址、标题、分数、通过数、提交数、通过率
3.详情页信息抽取
4.信息存储

```python
#1.构建两种网页
urls = [("https://pintia.cn/problem-sets/1111652100718116864/problems/type/6","函数题"),
        ("https://pintia.cn/problem-sets/1111652100718116864/problems/type/7","编程题")]
from selenium import webdriver
import re,time
driver=webdriver.Chrome()

def get_html(url):
    driver.get(url)
    time.sleep(5)
    html = driver.page_source
    return html

#2.正则信息抽取
def extract_info(html,category):
    import re
    results=[]#[[a,b,c,d,e,f,g],[]]
    reobj = re.compile(r'<tr><td class="answerIcon_1du7d"></td><td>([\d\D]*?)</td><td><a href="([\d\D]*?)" class="">([\d\D]*?)</a></td><td>([\d\D]*?)</td><td>([\d\D]*?)</td><td>([\d\D]*?)</td><td>([\d\D]*?)</td></tr>')
    for match in reobj.finditer(html):
        a=match.group(1)#标号
        b=match.group(2)#网址/problem-sets/1111652100718116864/problems/1111653039147499520
        b= "https://pintia.cn" + b
        c=match.group(3)#标题
        d=match.group(4)#分数
        e=match.group(5)#通过数
        f=match.group(6)#提交数
        g=match.group(7)#通过率
        results.append([a,b,c,d,e,f,g,category])
    return results

results = []#所有信息的存放
for t in urls:
    url , category = t
    html = get_html(url)
    li  = extract_info(html,category)
    for elem in li:
        results.append(elem)
    
print(results)
```

```python
#3.详情页信息抽取
def get_content(html):

    result = []
    reobj = re.compile(r'返回[\d\D]*?(<div class="ques-view">[\d\D]*</pre>[\d\D]*?</div>)[\d\D]*?陈春')
    for match in reobj.finditer(html):
        result.append(match.group(1))
    return result    

final_results = []
for li in results:
    url = li[1]
    html = get_html(url)

    content = get_content(html)
    li.append(content)
    final_results.append(li)

print(final_results)
```
